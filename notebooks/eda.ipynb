```python
df_encoded = df.copy()
df_encoded = pd.get_dummies(df_encoded, columns=["Occupation", "City_Tier"], drop_first=True)

plt.figure(figsize=(16, 12))
sns.heatmap(df_encoded.corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

Feature Engineering

```python
df = pd.get_dummies(df, columns=["Occupation", "City_Tier"], drop_first=True)
```

```python
df['Savings_Ratio'] = df['Disposable_Income'] / df['Income']
```

```python
essential_expenses = ["Rent", "Loan_Repayment", "Insurance", "Groceries", "Utilities", "Healthcare", "Education"]
non_essential_expenses = ["Transport", "Eating_Out", "Entertainment", "Miscellaneous"]

df["Essential_Expenses"] = df[essential_expenses].sum(axis=1)
df["Non_Essential_Expenses"] = df[non_essential_expenses].sum(axis=1)
```
